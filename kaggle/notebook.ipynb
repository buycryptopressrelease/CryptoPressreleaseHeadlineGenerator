```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/headline_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['headline_quality_score','seo_optimization_score','crypto_keywords_score','click_through_potential_score','media_distribution_score']
signal_avgs = df[signal_cols].mean()
colors = ['#f97316','#3b82f6','#10b981','#ef4444','#8b5cf6']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('CryptoPressRelease - Average Headline Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Headline Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=75, color='green', linestyle='--', alpha=0.5, label='Healthy threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
type_counts = df['announcement_type'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(type_counts.index, type_counts.values, color=colors[:len(type_counts)])
plt.title('CryptoPressRelease - Announcement Type Distribution', fontsize=14, fontweight='bold')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
platform_cols = ['google_news_score','coindesk_score','cointelegraph_score','pr_newswire_score']
platform_avgs = df[platform_cols].mean()
plt.figure(figsize=(8,5))
plt.bar(platform_avgs.index, platform_avgs.values, color=colors[:4])
plt.title('CryptoPressRelease - Platform Visibility Scores', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
